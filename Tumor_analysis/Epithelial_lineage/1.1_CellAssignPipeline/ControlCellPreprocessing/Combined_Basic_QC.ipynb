```python
import numpy as np
import pandas as pd
import scanpy as sc
import anndata
import matplotlib.pyplot as plt
import anndata2ri
import numpy as np

import rpy2.rinterface_lib.callbacks
import logging
import subprocess


ctrl_samples = {
}

ctrl_samples_save_files = {
}

```

```python
combined = sc.read_h5ad('data/Seurat2AdataIntermediateFiles/combined_ctrl.h5ad')
```

```python
combined.X = combined.raw.X
```

```python
print(combined.X)
```

```python
combined.obs
```

```python
sc.pp.filter_cells(combined ,min_genes=200)
sc.pp.filter_cells(combined ,max_genes=7500)
sc.pp.filter_genes(combined, min_cells=3)
```

```python
sc.external.pp.scrublet(combined)
combined = combined[~combined.obs["predicted_doublet"], :]
```

```python
combined.var['mt'] = combined.var_names.str.startswith('MT-')  

sc.pp.calculate_qc_metrics(combined, qc_vars=['mt'], percent_top=None, log1p=False, inplace=True)
    
combined = combined[combined.obs.pct_counts_mt < 10, :]
```

```python
all_data = pd.read_csv("data/raw_data/Epithelial_Ctrl_metadata-2.csv")

all_data["Indexes"] = all_data.apply(lambda x: x["barcode_orig"].strip("_6"), axis = 1)
cell_type_dict = dict(zip(all_data["Indexes"], all_data["cell_type_fine"]))
combined.obs["cell_type_fine"] = pd.Series(cell_type_dict)
combined.obs['cell_type_fine']= combined.obs['cell_type_fine'].fillna("NA")
```

```python
combined.var = combined.var[["features", "n_cells", "n_cells_by_counts", "mean_counts", 
                            "pct_dropout_by_counts", "total_counts"]].rename_axis("genes")
```

```python
combined.__dict__['_raw'].__dict__['_var'] = combined.__dict__['_raw'].__dict__['_var'].rename(columns={'_index': 'features'})


```

```python
combined.write_h5ad("data/basicQCAllCtrl.h5ad")
```
