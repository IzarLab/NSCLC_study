```python
import numpy as np
import pandas as pd
import scanpy as sc
import anndata
import matplotlib.pyplot as plt
import anndata2ri
import numpy as np

import rpy2.rinterface_lib.callbacks
import logging
import subprocess


ctrl_samples = {
}

ctrl_samples_save_files = {
}
sample_numbers = [1, 2,3,4,5,7]
```

```python
for i in sample_numbers:
    print("data/Seurat2AdataIntermediateFiles/C5%s_ctrl.h5ad"%i)
    print("ctrl_c5%s"%i)
    temp_obs = sc.read_h5ad("data/Seurat2AdataIntermediateFiles//C5%s_ctrl.h5ad"%i).obs #load Seurat Object Information
    temp_ctrl = sc.AnnData(pd.read_csv("data/Seurat2AdataIntermediateFiles/ctrl_c5%s.csv"%i).T) #The CSV is the rawcounts file
    save_file = 'data/SampleEpithelialCellsOnly/ctrl_c5%i.h5ad'%i
    temp_ctrl.obs = temp_obs
    ctrl_samples["ctrl_c5%s"%i] = temp_ctrl
    ctrl_samples_save_files["ctrl_c5%s"%i] = save_file
    all_data = pd.read_csv("data/raw_data/Epithelial_Ctrl_metadata-2.csv")
    all_data = all_data[all_data["orig.ident"] == "C5%sctr_3p"%i]
    print(len(all_data))
    print(len(ctrl_samples['ctrl_c5%s'%i].obs))
    all_data["Indexes"] = all_data.apply(lambda x: x["barcode_orig"].strip("_6"), axis = 1)
    cell_type_dict = dict(zip(all_data["Indexes"], all_data["cell_type_fine"]))
    ctrl_samples['ctrl_c5%s'%i].obs["cell_type_fine"] = pd.Series(cell_type_dict)
    ctrl_samples['ctrl_c5%s'%i] = ctrl_samples['ctrl_c5%s'%i][ctrl_samples['ctrl_c5%s'%i].obs['cell_type_fine'].notnull(),:]
```

```python
#Filtering based on genes per cell/min cells each gene is expressed

for x in ctrl_samples:
    sc.pp.filter_cells(ctrl_samples[x], min_genes=200)
    sc.pp.filter_cells(ctrl_samples[x], max_genes=7500)
    sc.pp.filter_genes(ctrl_samples[x], min_cells=3)
    

ctrl_samples['ctrl_c51']
```

```python
# annotate the group of mitochondrial genes as 'mt'
#filtering cell with >5 mitochondrial genes

for x in ctrl_samples:

    ctrl_samples[x].var['mt'] = ctrl_samples[x].var_names.str.startswith('MT-')  

    sc.pp.calculate_qc_metrics(ctrl_samples[x], qc_vars=['mt'], percent_top=None, log1p=False, inplace=True)
    
    ctrl_samples[x] = ctrl_samples[x][ctrl_samples[x].obs.pct_counts_mt < 10, :]
    
    
```

```python
for x in ctrl_samples:
    sc.external.pp.scrublet(ctrl_samples[x])
    ctrl_samples[x] = ctrl_samples[x][~ctrl_samples[x].obs["predicted_doublet"], :]

```

```python
#normalize to sum to 10000 per cell, compute size factor
for x in ctrl_samples:
    sc.pp.normalize_total(ctrl_samples[x], target_sum=1e4)
    lib_size = ctrl_samples[x].X.sum(1)
    ctrl_samples[x].obs["size_factor"] = lib_size / np.mean(lib_size)
    ctrl_samples[x] = ctrl_samples[x].copy()
```

```python
#Read All Signatures, Store as Dictionary
sigs_excel = "data/raw_data/Epithelial_lung_signature.xlsx"
covid_sigs_df = pd.read_excel(sigs_excel).T
covid_sigs = {}
for i,row in covid_sigs_df.fillna(0).iterrows():
    sig_set = []
    for sig in row:
        if sig != 0:
            sig_set.append(sig.strip("\xa0").strip())
    covid_sigs[i] = list(set(sig_set))
covid_sigs

laughney_sigs_df = pd.read_excel(sigs_excel, sheet_name='Laughney_signature').T
laughney_sigs = {}
for i,row in laughney_sigs_df.fillna(0).iterrows():
    sig_set = []
    for sig in row:
        if sig != 0:
            sig_set.append(str(sig).strip("\xa0").strip())
    laughney_sigs[i] = list(set(sig_set))
laughney_sigs


epi_sigs_df = pd.read_excel(sigs_excel, sheet_name='Epithelial_lung').T
epi_sigs = {}
for i,row in epi_sigs_df.fillna(0).iterrows():
    sig_set = []
    for sig in row:
        if sig != 0:
            sig_set.append(str(sig).strip("\xa0").strip())
    epi_sigs[i] = list(set(sig_set))

all_sigs = {
    "covid": covid_sigs,
    "laughney": laughney_sigs,
    "epi": epi_sigs
}

def create_marker_gene_matrix(sigs_dict):
    all_genes = []
    for x in sigs_dict:
        all_genes.extend(sigs_dict[x])
    all_genes = list(set(all_genes))
    df = []
    for x in sigs_dict:
        df_col = []
        for y in all_genes:
            df_col.append(int( y in sigs_dict[x]))
        df.append(df_col)
    return pd.DataFrame(df, index =list(sigs_dict.keys()), columns = all_genes).T
create_marker_gene_matrix(covid_sigs)


def get_cellassign_sample(sample, matrix):
    gene_matrix =  create_marker_gene_matrix(matrix)
    sample= sample[:,sample.var.index.isin(set(gene_matrix.index))]
    sample = sample.copy()
    return sample

for i in sample_numbers:
    for y in all_sigs:
        samp = get_cellassign_sample(ctrl_samples["ctrl_c5%s"%i], all_sigs[y])
        save_file = 'data/nonLoggedCellAssignInput/epi_only_ctrl_c5%s_'%i+y+".h5ad"
        samp = get_cellassign_sample(ctrl_samples['ctrl_c5%s'%i], all_sigs[y])

        samp.write(save_file)
```

```python
#log data and add size factor
       
for x in ctrl_samples:
    
    
    sc.pp.log1p(ctrl_samples[x])
```

```python
for x in ctrl_samples:

    sc.tl.pca(ctrl_samples[x], svd_solver='arpack')
    sc.pl.pca_variance_ratio(ctrl_samples[x], log=True)
```

```python
#computing neighbors
for x in ctrl_samples:
    sc.pp.neighbors(ctrl_samples[x], n_neighbors=10, n_pcs=40)


#compute leiden res = 1
for x in ctrl_samples:
    sc.tl.leiden(ctrl_samples[x])
```
