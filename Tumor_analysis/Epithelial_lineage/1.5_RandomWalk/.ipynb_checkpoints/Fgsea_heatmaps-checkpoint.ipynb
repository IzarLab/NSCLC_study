```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
pathway1 = pd.read_csv("fgsea/NES_scores.csv").set_index("pathway").drop("Unnamed: 0", axis = 1).fillna(0)
```

```python
pathway = pd.read_csv("fgsea_mixed_lineage/NES_scores.csv").set_index("pathway").drop("Unnamed: 0", axis = 1).fillna(0)
```

```python
pathway
```

```python
plt.figure(figsize=(3, 5), dpi=250)  # Increased DPI for better resolution
ax = sns.heatmap(pathway1.T, annot=True, fmt=".1f", cmap="coolwarm", center=0, 
                 linewidths=.5, annot_kws={"size": 12 , "color": "black"})  # Increased text size
plt.title("Heatmap of Mutation Pathways Across Different Clusters")
plt.ylabel("Lineage Clusters")
plt.xlabel("Genotypes")

# Adjusting tick labels for better readability
plt.xticks(rotation=90)  # Making x-axis labels horizontal

# Adding "NES" label on top of the colorbar
colorbar = ax.collections[0].colorbar
colorbar.set_label('NES', labelpad=10)

plt.savefig('fgsea/heatmatpNES.pdf', bbox_inches='tight')  # Save to PDF

plt.show()
```

```python
plt.figure(figsize=(3, 5), dpi=250)  # Increased DPI for better resolution
ax = sns.heatmap(pathway.T, annot=True, fmt=".1f", cmap="coolwarm",center=0, 
                 linewidths=.5, annot_kws={"size": 12,"color": "black"})  # Increased text size
plt.title("Heatmap of Mutation Pathways Across Different Clusters")


# Labeling axes
plt.ylabel("Mixed Lineage Clusters")
plt.xlabel("Genotypes")

# Adjusting tick labels for better readability
plt.xticks(rotation=90)  # Making x-axis labels horizontal

# Adding "NES" label on top of the colorbar
colorbar = ax.collections[0].colorbar
colorbar.set_label('NES', labelpad=10)
plt.savefig('fgsea_mixed_lineage/heatmatpNES_ML.pdf', bbox_inches='tight')  # Save to PDF


plt.show()
```
